```
courses = []
choice = 'random'

def line():
    print('---------------------------')

def show_menu():
    line()
    print('MENU')
    line()
    print('1. Add Semester Course')
    print('2. Unenroll a Course')
    print('3. View Courses')
    print('4. Exit')
    line()
    choice = input('Enter your choice: ')
    return choice

while choice != '4':
    choice = show_menu()
    if choice == '1':
        course = input('Enter course name you want to add? ')
        courses.append(course)
        print('Course', course, 'added successfully!')
    elif choice == '2':
        if courses != []:
          print('List of courses:')
          for course in courses:
              print('●', course)
          course = input('Enter which course you want to unenroll from: ')
          if course in courses:
              courses.remove(course)
              print('Removed course', course)
          else:
              print('Could not find course', course)
        else:
              print('There is no course you have added yet!')
    elif choice == '3':
        if courses != []:
          print('List of courses:')
          for course in courses:
              print('●', course)3
        else:
              print('There is no course you have added yet!')
    elif choice == '4':
        print('Goodbye!')
    else:
        print('Please enter one of 1, 2, 3 or 4')
```
